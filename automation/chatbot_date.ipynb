```python
import pandas as pd
```

```python
df = pd.read_excel('chatbot_dates.xlsx', sheet=1)
df.head()
```

```python
app_date = pd.to_datetime(df['app_datetime'])
app_date = app_date + pd.DateOffset(days=16)
df['app_datetime'] = app_date.dt.strftime('%m/%d/%Y %I:%M %p')
df.head()
```

```python
df.to_excel('chatbot2.xlsx',index=False)
```
